# Reverse Geocoding -  Latitude/ Longitude to City/Country - Python and Pandas

```python
import pandas as pd

df = pd.read_csv(f'../data/earthquakes_1965_2016_database.csv.zip')
df
```

## Step 2: Convert latitude and longtitude to city/country with geopy

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="http")
location = geolocator.reverse("51.5074, 0.1278")

print(location.address)
```

```python
location.raw.get('address').get('city')
```

```python
location.raw
```

## Step 3: Reverse Geocoding with geocoder

```python
import geocoder

g = geocoder.osm([51.5074, 0.1278], method='reverse')
g.json['city']
```

```python
g.json
```

## Pandas

```python
import geocoder

def geo_rev(x):
    g = geocoder.osm([x.Latitude, x.Longitude], method='reverse').json
    if g:
        return g.get('country')
    else:
        return 'no country'

df[['Latitude', 'Longitude']].tail().apply(geo_rev, axis=1)
```
