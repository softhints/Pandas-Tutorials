# How to Sort Pivot Table by the Margin All in Pandas

```python
import pandas as pd
df = pd.DataFrame({'foo': ['one', 'one', 'one', 'two', 'two',
                           'two'],
                   'bar': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'baz': [1, 2, 3, 4, 5, 6],
                   'zoo': ['x', 'y', 'z', 'q', 'w', 't']})
```

```python
df
```

```python
df.pivot_table(index='foo', columns='bar', values='baz', aggfunc=['sum'], margins=True)
```
