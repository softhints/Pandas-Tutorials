# How to Search and Download Kaggle Dataset to Pandas DataFrame

https://datascientyst.com/search-download-kaggle-dataset-pandas-dataframe/

```python
import kaggle

datafile = ['database.csv', 'medium_data.csv', 'StudentsPerformance.csv', 'internshala_dataset_raw.csv'][0]

kaggle.api.authenticate()
kaggle.api.dataset_download_file('dhairyatalsania/internshalaindia-raw-dataset', file_name=datafile,  path='data/')
```

```python
import pandas as pd

datafile = ['database.csv', 'medium_data.csv', 'StudentsPerformance.csv', 'internshala_dataset_raw.csv'][1]
try:
    df = pd.read_csv(f'data/{datafile}')
except:
    df = pd.read_csv(f'data/{datafile}.zip')
df
```
